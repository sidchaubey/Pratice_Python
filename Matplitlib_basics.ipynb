```
9import matplotlib.pyplot as plt
import numpy as np
a = np.random.randint(1,10,10)
plt.plot(a)
plt.show()
# here on x axis as we've not expliticly mentioned indices therefore it takes indices as x values
```

```
a = [1,5,6,3]
plt.plot(a)
plt.title("a vs index")
plt.xlabel('Indices')
plt.ylabel("Numbers")
plt.show()
```

```
# A Grid helpes us identify values easily
a = [1,5,6,3]
b = [5,3,4,5]
plt.plot(a,b,"o")
plt.title("a vs b")
plt.xlabel("a")
plt.ylabel("b")
plt.xlim(0,10) # Setting x  min and max scale
plt.ylim(0,10)# Setting y  min and max scale
plt.grid() # Setting grid lines
plt.show()
```

```
# If I want to plot multiple things on same plot
from math import pi
a = np.arange(-2*pi,2*pi,0.01)
b = np.sin(a)
c = np.cos(a)
# d = np.tanh(a)
plt.plot(a,b,label="sin",color="r")
plt.plot(a,c,label="cos",color = "b")
# plt.plot(a,d,label="tan",color = "g")
plt.legend()
plt.show()
```

```
# If I want to plot multiple things on same plot
from math import pi
a = np.arange(-pi,pi,0.01)
d = np.tanh(a)
# e = np.arctanh(a) # Inverse hyperbolic function
plt.plot(a,d,label="tanh",color = "g")
# plt.plot(a,e,label="arctanh",color = "b")
plt.legend()
plt.show()
```

```
a = np.arange(10)
b = np.exp(a)
c = np.exp(np.arange(1,11))
plt.plot(a,b,'b--',label="b")
plt.plot(a,c,'g-*',label="c")
plt.legend()
plt.show()
```

```
a = np.arange(10)
b = np.exp(a)
c = np.exp(np.arange(1,11))
plt.plot(a,b,'g*',label="b",linewidth=0.5)
plt.plot(a,c,'r^',label="c",linewidth=0.5)
plt.legend()
plt.show()
```

```
# Working through Multiple figures 
def fig(t):
  return np.exp(t)*np.cos(t)

a = np.arange(1,10)
b = np.arange(11,20)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(a,fig(a),"b--")

plt.subplot(212)
plt.grid()
plt.plot(a,fig(b),"b--")
plt.show()
```

```
from scipy.stats import poisson
def poissio(a):
  return poisson(a)
b = poissio(np.arange(10)).cdf(0.1)
print(b.cdf(x=0.1))
```
