# CDF(Cumulative Distribution Frequency):
* ***pdf*** basically tells us the count of occurance of a particular point in the distribution.***pdf*** is basically the derivative of ***cdf***.
* ***cdf*** basically gives us an idea about total how many points are below a particular point in a distribution. ***cdf***  is basically the integration of area under the ***pdf***.

```
# Code 
iris_setosa = df[df["species"]=="setosa"]
iris_versicolor = df[df["species"]=="versicolor"]
iris_virginica = df[df["species"]=="virginica"]

# plot CDF of petal_length
counts, bin_edges = np.histogram(iris_setosa["petal_length"],bins=10
                                 ,density=True)
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show();
```

```
# CDF setosa in case of petal_length

count, bin_size = np.histogram(iris_setosa["petal_length"],bins=10,
                               density=True)
pdf = count/sum(count)
# print(pdf)
# print(bin_size)

cdf = np.cumsum(pdf)
plt.plot(bin_size[1:],pdf,label="setosa_pdf")
plt.plot(bin_size[1:],cdf,label="setosa_cdf");

count, bin_size = np.histogram(iris_versicolor["petal_length"],bins=10,
                               density=True)
pdf = count/sum(count)
# print(pdf)
# print(bin_size)

cdf = np.cumsum(pdf)
plt.plot(bin_size[1:],pdf,label="versicolor_pdf")
plt.plot(bin_size[1:],cdf,label="versicolor_cdf");

count, bin_size = np.histogram(iris_virginica["petal_length"],bins=10,
                               density=True)
pdf = count/sum(count)
# print(pdf)
# print(bin_size)

cdf = np.cumsum(pdf)
plt.plot(bin_size[1:],pdf,label="virginica_pdf")
plt.plot(bin_size[1:],cdf,label="virginica_cdf");

plt.legend()
plt.show()
```

# Mean, Variance and Std-dev




```
# mean, Variance, standard-dev
print("Means: ")
print(np.mean(iris_setosa["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
print(np.mean(iris_virginica["petal_length"]))
# Mean with an outlier
print(np.mean(np.append(iris_setosa["petal_length"],50)))
# Std-dev
print(" ")
print(" ")
print("Standard-dev")
print(np.std(iris_setosa["petal_length"]))
print(np.std(iris_versicolor["petal_length"]))
print(np.std(iris_virginica["petal_length"]))
# std with an outlier
print(np.std(np.append(iris_setosa["petal_length"],50)))
```

```
"""Note: 
mean and standard-deviation tends to get affected 
a lot if there is even a single outlier present. 
As we can see in above example
"""
```

# Median, Percentile, Quantile, IQR, MAD

```
print("Medians: ")
print(np.median(iris_setosa["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))
print(np.median(iris_virginica["petal_length"]))
# Median with an outlier
print(np.median(np.append(iris_setosa["petal_length"],50)))
# Percentile
print(" ")
print(" ")
print("Percentile")
print(np.percentile(iris_setosa["petal_length"],10))
print(np.percentile(iris_versicolor["petal_length"],10))
print(np.percentile(iris_virginica["petal_length"],10))
print(" ")
print(" ")
# Quantiles
print("Quantiles: ")
print(np.percentile(iris_setosa["petal_length"],np.arange(0,101,25)))
print(np.percentile(iris_versicolor["petal_length"],np.arange(0,101,25)))
print(np.percentile(iris_virginica["petal_length"],np.arange(0,101,25)))
```

* Note:
1. Mean is highly affected even if there is only a single large outlier value. Whereas in case of median is not affected if there is only single outlier values.
2. Median is affected only if there are more than 50% points are outlier.

```
from scipy.stats import median_absolute_deviation
median_abd_setosa = median_absolute_deviation(iris_setosa["sepal_length"])
print("Setosa median_absolute_deviation: ",median_abd_setosa)
median_abd_virginica = median_absolute_deviation(iris_virginica["sepal_length"])
print("virginica median_absolute_deviation: ",median_abd_virginica)
median_abd_versicolor = median_absolute_deviation(iris_versicolor["sepal_length"])
print("Versicolor median_absolute_deviation: ",median_abd_setosa)
```

```
from statsmodels import robust
print(robust.mad(iris_setosa["sepal_length"]))
print(robust.mad(iris_versicolor["sepal_length"]))
print(robust.mad(iris_virginica["sepal_length"]))
```

* Median <---> Mean
* MAD(median absolute Deviation) <---> Std-dev(standard deviation)
* IQR stands for inter quartile range. IQR is computed by substracting 75th percentile - 25th percentile.
