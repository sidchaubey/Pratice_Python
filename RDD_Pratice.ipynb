```
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName("RDD")
sc = SparkContext.getOrCreate(conf = conf)
```

```
rdd = sc.textFile("/FileStore/tables/map.txt")
rdd.collect()
```

```
rdd1 = rdd.map(lambda x: x.split(" "))
rdd1.collect()
```

```
rdd2 = rdd1.flatMap(lambda x: x)
rdd2.collect()
```

```
i = 1;
for j in rdd2.collect():
  print(f"Element: {i}, Value: {j}")
  i +=1
```

```
rdd3 = rdd1.flatMap(lambda x: x).map(lambda x: (x,1))
rdd3.collect()
```

```
def removing_special_symbol(x):
  if x[-1]=="?":
    return x[:-1]
  return x

rdd4 = rdd2.map(removing_special_symbol).map(lambda x: (x, 1))
rdd4.collect()
```

```
rdd5 = rdd4.groupByKey()
rdd5.mapValues(list).collect()
```

```
rdd6 = rdd4.reduceByKey(lambda x,y: x+y)
rdd6.collect()
```

```
rdd2.count()
```

```
rdd2.countByKey()
```

```
conf = SparkConf().setAppName("filter")
sc = SparkContext.getOrCreate(conf = conf)
```

```
rdd = sc.textFile("/FileStore/tables/map.txt")
rdd.collect()
```

```
rdd1 = rdd.flatMap(lambda x: x.split(" "))
rdd1.collect()
```

```
def remove_questionmark(x):
  if x[-1]=="?":
    return x[:-1]
  return x
```

```
rdd2 = rdd1.map(remove_questionmark)
rdd2.collect()
```

```
# groupByKey

rdd3 = rdd2.map(lambda x: (x, 1))
rdd3.collect()
```

```
rdd4 = rdd3.groupByKey()
rdd4.mapValues(list).collect()
```

```
# Reduce By Key

rdd3.collect()
rdd5 = rdd3.reduceByKey(lambda x,y: x+y)
 
for i in rdd5.collect():
  print("Key: {},\n Number of times Occuring: {}\n\n\n".format(i[0], i[1]))
```

```
# Filter:
rdd1.collect()
def remove_wordwith_questionmark(x):
  if x[-1]!="?":
    return x

rdd6 = rdd1.filter(remove_wordwith_questionmark)
rdd6.collect()
```

```
print(len(rdd2.collect()))
rdd7 = rdd2.distinct()
len(rdd7.collect())
```
